```python
img = cv2.imread(r'C:\Users\junio\Desktop\OpenCV-master-v6_00\DATA\crossword.jpg', 0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 170, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15,8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.7, src2=th2, beta=0.3, gamma=0)
show_pic(blended)
```
