```python
blank_img = np.zeros((512,512,3), dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, (100,100), (400,400), color=(0,255,0), thickness= -1)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, (250,250), radius=50, color=(0,0,250), thickness=-1)
plt.imshow(blank_img)
```

```python
img = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(img)
```

```python
cv2.putText(img, text='Test', org=(100,255), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=4, color=(255, 222, 173), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(img)
```

```python
punkt = np.array([[100,150], [100,270], [450, 270], [450,150]], np.int32)
punkt.shape
```

```python
pts = punkt.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(img, [pts], isClosed= True, color=(125, 255, 255), thickness=5, lineType=cv2.FONT_HERSHEY_COMPLEX )
plt.imshow(img)
```
